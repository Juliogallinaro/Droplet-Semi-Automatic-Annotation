# Dataset Merge and CVAT Export

Merge multiple datasets generated by the pipeline into a single `dataset_merged` folder and export CVAT ZIP files.

```
dataset_merged/
├── images/train/
├── images/val/
├── labels/train/
├── labels/val/
├── dataset.yaml
├── cvat_train.zip
└── cvat_val.zip
```

## 2. Imports

````python
from pathlib import Path
from droplet_annotation import export_cvat_archives, merge_datasets, write_dataset_yaml
````

## 3. Settings  <- **edit here**

````python
# Parameters
DATASET_DIRS = [
    "dataset_1",
    "dataset_2",

]
OUTPUT_DIR = "dataset_merged"
CLASS_NAME = "droplet"

print("Configuration loaded.")
````

## 4. Merge the Datasets

````python
result = merge_datasets(DATASET_DIRS, OUTPUT_DIR)
yaml_path = write_dataset_yaml(OUTPUT_DIR, CLASS_NAME)

out = Path(OUTPUT_DIR)

print(f"Copied files     : {result.copied}")
print(f"Renamed conflicts: {result.conflicts}")
print()
print(f"train : {result.train_images} images")
print(f"val   : {result.val_images} images")
print(f"dataset.yaml saved to: {yaml_path.resolve()}")
````

## 5. Export CVAT Annotation Archives

````python
n_train, n_val = export_cvat_archives(OUTPUT_DIR, CLASS_NAME)

print(f"cvat_train.zip : {n_train} annotations -> obj_train_data/")
print(f"cvat_val.zip   : {n_val} annotations -> obj_valid_data/")
print()
print("In CVAT, create one task per split (train and val):")
print("1. Create Task and upload the images for that split.")
print("2. Actions -> Upload annotations -> YOLO 1.1 -> matching ZIP file.")
````
